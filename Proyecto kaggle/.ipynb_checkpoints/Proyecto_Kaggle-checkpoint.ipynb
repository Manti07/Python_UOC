```python
import pandas as pd
import numpy as np
import re
import datetime


Book_dataset= pd.read_csv('/home/manti/uoc/Python_UOC/Proyecto kaggle/BooksDataset.csv')

Clean_Book_dataset = Book_dataset.dropna(subset=['Category'])
```

```python
Clean_Book_dataset.loc[:,'Clean Price'] = Clean_Book_dataset['Price'].apply(lambda x: re.findall('[0-9]*\.[0-9]*', x)[0])
Clean_Book_dataset['Clean Price'][0]
```

```python
from datetime import datetime
def conversor_datetime(x):
    numeros = re.findall('[a-zA-Z]*\,\s([a-zA-Z]*)\s([0-9]*)\,\s([0-9]*)', x)
    
    if len(numeros)<1:
        old_datetime_object= datetime.strptime(x, "%B %Y")
        return old_datetime_object

    date_time_str = numeros[0][0] + ' ' + numeros[0][1] + ' '+ numeros[0][2]
    datetime_object= datetime.strptime(date_time_str, "%B %d %Y")

    return datetime_object
    


# Clean_Book_dataset['Datetime date']= (Clean_Book_dataset['Publish Date'].apply(lambda x: re.findall('[a-zA-Z]*\,\s([a-zA-Z]*)\s([0-9]*)\,\s([0-9]*)', x)))
Clean_Book_dataset['Datetime']= (Clean_Book_dataset['Publish Date'].apply(conversor_datetime))
Clean_Book_dataset
```
